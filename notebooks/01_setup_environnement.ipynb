```python
#@title ⚙️ Configuration { display-mode: "form" }

import os
import subprocess
from pathlib import Path

# Monter Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Chemins du projet
PROJECT_DIR = Path("/content/drive/MyDrive/dubbing_project")
INPUT_DIR = PROJECT_DIR / "inputs"
OUTPUT_DIR = PROJECT_DIR / "outputs"
TEMP_DIR = PROJECT_DIR / "temp"

# Créer les dossiers si nécessaire
for d in [INPUT_DIR, OUTPUT_DIR, TEMP_DIR]:
    d.mkdir(parents=True, exist_ok=True)

print("📁 Structure du projet :")
print(f"   Inputs  : {INPUT_DIR}")
print(f"   Outputs : {OUTPUT_DIR}")
print(f"   Temp    : {TEMP_DIR}")
```

```python
#@title 🎬 Module Extraction { display-mode: "form" }

class ExtractionModule:
    """
    Module d'extraction audio/vidéo utilisant FFmpeg.

    Fonctionnalités :
    - Extraire l'audio en WAV (16kHz, mono) pour Whisper
    - Créer une vidéo muette pour la fusion finale
    - Récupérer les métadonnées de la vidéo
    """

    def __init__(self, temp_dir: Path):
        """
        Initialise le module.

        Args:
            temp_dir: Dossier pour les fichiers temporaires
        """
        self.temp_dir = Path(temp_dir)
        self.temp_dir.mkdir(parents=True, exist_ok=True)

        # Vérifier que FFmpeg est disponible
        self._check_ffmpeg()

    def _check_ffmpeg(self):
        """Vérifie que FFmpeg est installé."""
        try:
            result = subprocess.run(
                ['ffmpeg', '-version'],
                capture_output=True,
                text=True
            )
            print("✅ FFmpeg disponible")
        except FileNotFoundError:
            raise RuntimeError("❌ FFmpeg non trouvé !")

    def get_video_info(self, video_path: str) -> dict:
        """
        Récupère les métadonnées d'une vidéo.

        Args:
            video_path: Chemin de la vidéo

        Returns:
            dict avec durée, fps, résolution, etc.
        """
        cmd = [
            'ffprobe',
            '-v', 'quiet',
            '-print_format', 'json',
            '-show_format',
            '-show_streams',
            str(video_path)
        ]

        result = subprocess.run(cmd, capture_output=True, text=True)

        import json
        data = json.loads(result.stdout)

        # Extraire les infos utiles
        video_stream = None
        audio_stream = None

        for stream in data.get('streams', []):
            if stream['codec_type'] == 'video' and video_stream is None:
                video_stream = stream
            elif stream['codec_type'] == 'audio' and audio_stream is None:
                audio_stream = stream

        info = {
            'duration': float(data['format'].get('duration', 0)),
            'size_mb': int(data['format'].get('size', 0)) / (1024 * 1024),
        }

        if video_stream:
            info['width'] = video_stream.get('width')
            info['height'] = video_stream.get('height')
            # Calculer FPS
            fps_str = video_stream.get('r_frame_rate', '25/1')
            num, den = map(int, fps_str.split('/'))
            info['fps'] = num / den if den else 25

        if audio_stream:
            info['audio_sample_rate'] = int(audio_stream.get('sample_rate', 44100))
            info['audio_channels'] = audio_stream.get('channels', 2)

        return info

    def extract_audio(self, video_path: str, output_path: str = None) -> str:
        """
        Extrait l'audio d'une vidéo en WAV (16kHz mono).

        Format optimisé pour Whisper (ASR).

        Args:
            video_path: Chemin de la vidéo source
            output_path: Chemin de sortie (optionnel)

        Returns:
            Chemin du fichier audio extrait
        """
        video_path = Path(video_path)

        if output_path is None:
            output_path = self.temp_dir / f"{video_path.stem}_audio.wav"

        output_path = Path(output_path)

        print(f"🔊 Extraction audio : {video_path.name}")

        cmd = [
            'ffmpeg',
            '-i', str(video_path),
            '-vn',                    # Pas de vidéo
            '-acodec', 'pcm_s16le',   # Format WAV
            '-ar', '16000',           # 16kHz (optimal pour Whisper)
            '-ac', '1',               # Mono
            '-y',                     # Écraser si existe
            str(output_path)
        ]

        result = subprocess.run(cmd, capture_output=True, text=True)

        if result.returncode != 0:
            print(f"❌ Erreur FFmpeg : {result.stderr}")
            raise RuntimeError("Extraction audio échouée")

        print(f"   ✅ Audio extrait : {output_path.name}")
        return str(output_path)

    def extract_video_mute(self, video_path: str, output_path: str = None) -> str:
        """
        Crée une copie de la vidéo sans audio.

        Args:
            video_path: Chemin de la vidéo source
            output_path: Chemin de sortie (optionnel)

        Returns:
            Chemin de la vidéo muette
        """
        video_path = Path(video_path)

        if output_path is None:
            output_path = self.temp_dir / f"{video_path.stem}_mute.mp4"

        output_path = Path(output_path)

        print(f"📹 Extraction vidéo muette : {video_path.name}")

        cmd = [
            'ffmpeg',
            '-i', str(video_path),
            '-an',                    # Pas d'audio
            '-c:v', 'copy',           # Copie sans ré-encodage (rapide)
            '-y',
            str(output_path)
        ]

        result = subprocess.run(cmd, capture_output=True, text=True)

        if result.returncode != 0:
            print(f"❌ Erreur FFmpeg : {result.stderr}")
            raise RuntimeError("Extraction vidéo muette échouée")

        print(f"   ✅ Vidéo muette : {output_path.name}")
        return str(output_path)

    def process(self, video_path: str) -> dict:
        """
        Traitement complet : extrait audio + vidéo muette + métadonnées.

        Args:
            video_path: Chemin de la vidéo source

        Returns:
            dict avec tous les chemins et infos
        """
        print("=" * 50)
        print("🎬 EXTRACTION EN COURS")
        print("=" * 50)

        # 1. Métadonnées
        print("\n📊 Récupération des métadonnées...")
        info = self.get_video_info(video_path)
        print(f"   Durée     : {info['duration']:.1f} secondes")
        print(f"   Résolution: {info.get('width', '?')}x{info.get('height', '?')}")
        print(f"   FPS       : {info.get('fps', '?')}")

        # 2. Extraction audio
        print()
        audio_path = self.extract_audio(video_path)

        # 3. Vidéo muette
        print()
        video_mute_path = self.extract_video_mute(video_path)

        print()
        print("=" * 50)
        print("✅ EXTRACTION TERMINÉE")
        print("=" * 50)

        return {
            'video_original': str(video_path),
            'audio_path': audio_path,
            'video_mute_path': video_mute_path,
            'info': info
        }


# Instancier le module
extractor = ExtractionModule(temp_dir=TEMP_DIR)
print("\n✅ Module Extraction prêt !")
```

```python
#@title 🎤 Module ASR avec gestion mémoire { display-mode: "form" }

import torch
import gc
import whisper
from pathlib import Path

def clear_gpu_memory():
    """Libère la mémoire GPU."""
    gc.collect()
    torch.cuda.empty_cache()
    if torch.cuda.is_available():
        used = torch.cuda.memory_allocated(0) / 1e9
        print(f"📊 Mémoire GPU utilisée : {used:.1f} GB")

class ASRModule:
    """
    Module ASR avec gestion de mémoire.
    """

    def __init__(self, model_size: str = "large-v3"):
        self.model_size = model_size
        self.model = None

    def load(self):
        """Charge le modèle en mémoire."""
        if self.model is None:
            print(f"⏳ Chargement Whisper {self.model_size}...")
            clear_gpu_memory()
            self.model = whisper.load_model(self.model_size)
            print(f"✅ Whisper {self.model_size} chargé !")

    def unload(self):
        """Libère le modèle de la mémoire."""
        if self.model is not None:
            del self.model
            self.model = None
            clear_gpu_memory()
            print("🧹 Mémoire Whisper libérée")

    def transcribe(self, audio_path: str, language: str = "fr") -> dict:
        """Transcrit un fichier audio."""

        # Charger si pas déjà fait
        self.load()

        audio_path = Path(audio_path)
        print(f"🎤 Transcription : {audio_path.name}")
        print("   ⏳ En cours...")

        result = self.model.transcribe(
            str(audio_path),
            language=language,
            task="transcribe",
            verbose=False
        )

        texte_complet = result["text"].strip()

        segments = []
        for seg in result["segments"]:
            segments.append({
                "start": seg["start"],
                "end": seg["end"],
                "text": seg["text"].strip()
            })

        print()
        print("=" * 50)
        print("📝 TRANSCRIPTION")
        print("=" * 50)
        print(texte_complet)
        print("=" * 50)

        return {
            "text": texte_complet,
            "segments": segments,
            "language": language
        }

    def display_segments(self, segments: list):
        """Affiche les segments."""
        print("\n⏱️ SEGMENTS TEMPORELS :")
        print("-" * 50)
        for seg in segments:
            print(f"[{seg['start']:06.2f} → {seg['end']:06.2f}] {seg['text']}")
        print("-" * 50)


# Libérer toute mémoire existante d'abord
clear_gpu_memory()

# Créer le module (ne charge PAS encore le modèle)
asr = ASRModule(model_size="large-v3")

print("\n✅ Module ASR prêt (modèle sera chargé à la demande)")
```

```python
#@title 🌍 Module Traduction (NLLB) { display-mode: "form" }

import torch
import gc
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from pathlib import Path

def clear_gpu_memory():
    """Libère la mémoire GPU."""
    gc.collect()
    torch.cuda.empty_cache()
    if torch.cuda.is_available():
        used = torch.cuda.memory_allocated(0) / 1e9
        print(f"📊 Mémoire GPU utilisée : {used:.1f} GB")

class TraductionModule:
    """
    Module de traduction FR → Yoruba utilisant NLLB-200.

    NLLB (No Language Left Behind) est un modèle de Meta
    supportant 200 langues, dont plusieurs langues africaines.
    """

    def __init__(self, model_name: str = "facebook/nllb-200-1.3B"):
        """
        Initialise le module.

        Args:
            model_name: Nom du modèle NLLB
                - "facebook/nllb-200-distilled-600M" : Plus léger
                - "facebook/nllb-200-1.3B" : Bon compromis ✅
                - "facebook/nllb-200-3.3B" : Meilleur mais lourd
        """
        self.model_name = model_name
        self.model = None
        self.tokenizer = None

        # Codes langues NLLB
        self.lang_codes = {
            "francais": "fra_Latn",
            "yoruba": "yor_Latn",
            "fon": "fon_Latn"
        }

    def load(self):
        """Charge le modèle en mémoire."""
        if self.model is None:
            print(f"⏳ Chargement NLLB ({self.model_name.split('/')[-1]})...")
            print("   (Peut prendre 1-2 minutes)")

            clear_gpu_memory()

            self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
            self.model = AutoModelForSeq2SeqLM.from_pretrained(
                self.model_name,
                torch_dtype=torch.float16,  # Économise la mémoire
                device_map="auto"
            )

            print(f"✅ NLLB chargé !")
            clear_gpu_memory()

    def unload(self):
        """Libère le modèle de la mémoire."""
        if self.model is not None:
            del self.model
            del self.tokenizer
            self.model = None
            self.tokenizer = None
            clear_gpu_memory()
            print("🧹 Mémoire NLLB libérée")

    def traduire(self, texte: str, source: str = "francais", cible: str = "yoruba") -> str:
        """
        Traduit un texte.

        Args:
            texte: Texte à traduire
            source: Langue source ("francais")
            cible: Langue cible ("yoruba" ou "fon")

        Returns:
            Texte traduit
        """
        # Charger si nécessaire
        self.load()

        source_code = self.lang_codes[source]
        cible_code = self.lang_codes[cible]

        print(f"🌍 Traduction : {source} → {cible}")
        print(f"   Texte : {texte[:80]}...")
        print("   ⏳ En cours...")

        # Configurer la langue source
        self.tokenizer.src_lang = source_code

        # Tokenizer
        inputs = self.tokenizer(texte, return_tensors="pt").to(self.model.device)

        # Générer la traduction
        translated_tokens = self.model.generate(
            **inputs,
            forced_bos_token_id=self.tokenizer.convert_tokens_to_ids(cible_code),
            max_length=256,
            num_beams=5,
            repetition_penalty=1.2,      # Évite les répétitions
            no_repeat_ngram_size=3,       # Évite les boucles
            early_stopping=True
        )

        # Décoder
        traduction = self.tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

        print()
        print("=" * 50)
        print("📝 TRADUCTION")
        print("=" * 50)
        print(traduction)
        print("=" * 50)

        return traduction

    def traduire_segments(self, segments: list, source: str = "francais", cible: str = "yoruba") -> list:
        """
        Traduit une liste de segments.

        Args:
            segments: Liste de segments avec "text", "start", "end"
            source: Langue source
            cible: Langue cible

        Returns:
            Liste de segments traduits
        """
        # Charger si nécessaire
        self.load()

        print(f"🌍 Traduction de {len(segments)} segments...")
        print()

        segments_traduits = []

        for i, seg in enumerate(segments, 1):
            texte_original = seg["text"]

            print(f"[{i}/{len(segments)}] {texte_original[:50]}...")

            # Configurer la langue source
            self.tokenizer.src_lang = self.lang_codes[source]

            # Tokenizer
            inputs = self.tokenizer(texte_original, return_tensors="pt").to(self.model.device)

            # Générer
            translated_tokens = self.model.generate(
                **inputs,
                forced_bos_token_id=self.tokenizer.convert_tokens_to_ids(self.lang_codes[cible]),
                max_length=256,
                num_beams=5,
                repetition_penalty=1.2,
                no_repeat_ngram_size=3,
                early_stopping=True
            )

            traduction = self.tokenizer.decode(translated_tokens[0], skip_special_tokens=True)

            segments_traduits.append({
                "start": seg["start"],
                "end": seg["end"],
                "text_fr": texte_original,
                "text_yoruba": traduction
            })

            print(f"   → {traduction[:50]}...")
            print()

        return segments_traduits


# Libérer mémoire et créer le module
clear_gpu_memory()

traducteur = TraductionModule(model_name="facebook/nllb-200-1.3B")

print("\n✅ Module Traduction prêt !")
```

```python
#@title 🔊 Module TTS (MMS-TTS Yoruba) { display-mode: "form" }

import torch
import gc
from transformers import VitsModel, AutoTokenizer
import scipy.io.wavfile as wavfile
from pathlib import Path

def clear_gpu_memory():
    """Libère la mémoire GPU."""
    gc.collect()
    torch.cuda.empty_cache()
    if torch.cuda.is_available():
        used = torch.cuda.memory_allocated(0) / 1e9
        print(f"📊 Mémoire GPU utilisée : {used:.1f} GB")

class TTSModule:
    """
    Module de synthèse vocale (TTS) pour le Yoruba.

    Utilise MMS-TTS (Massively Multilingual Speech) de Meta,
    qui supporte plus de 1000 langues dont le Yoruba.

    Note : C'est le baseline. La qualité est "robotique".
    L'amélioration via fine-tuning VITS est la contribution
    principale du mémoire.
    """

    def __init__(self, model_name: str = "facebook/mms-tts-yor"):
        """
        Initialise le module TTS.

        Args:
            model_name: Modèle MMS-TTS pour Yoruba
        """
        self.model_name = model_name
        self.model = None
        self.tokenizer = None

    def load(self):
        """Charge le modèle en mémoire."""
        if self.model is None:
            print(f"⏳ Chargement MMS-TTS Yoruba...")

            clear_gpu_memory()

            self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
            self.model = VitsModel.from_pretrained(self.model_name)

            # Mettre sur GPU si disponible
            if torch.cuda.is_available():
                self.model = self.model.to("cuda")

            print(f"✅ MMS-TTS Yoruba chargé !")
            clear_gpu_memory()

    def unload(self):
        """Libère le modèle de la mémoire."""
        if self.model is not None:
            del self.model
            del self.tokenizer
            self.model = None
            self.tokenizer = None
            clear_gpu_memory()
            print("🧹 Mémoire TTS libérée")

    def synthetiser(self, texte: str, output_path: str = None) -> str:
        """
        Génère un audio à partir du texte Yoruba.

        Args:
            texte: Texte en Yoruba à synthétiser
            output_path: Chemin de sortie (optionnel)

        Returns:
            Chemin du fichier audio généré
        """
        # Charger si nécessaire
        self.load()

        print(f"🔊 Synthèse vocale...")
        print(f"   Texte : {texte[:80]}...")
        print("   ⏳ En cours...")

        # Tokenizer
        inputs = self.tokenizer(texte, return_tensors="pt")

        # Mettre sur GPU si disponible
        if torch.cuda.is_available():
            inputs = {k: v.to("cuda") for k, v in inputs.items()}

        # Générer l'audio
        with torch.no_grad():
            output = self.model(**inputs)

        # Récupérer le waveform
        waveform = output.waveform[0].cpu().numpy()

        # Sauvegarder
        if output_path is None:
            output_path = TEMP_DIR / "audio_yoruba.wav"

        output_path = Path(output_path)

        # Sample rate de MMS-TTS
        sample_rate = self.model.config.sampling_rate

        # Sauvegarder en WAV
        wavfile.write(str(output_path), sample_rate, waveform)

        # Calculer la durée
        duration = len(waveform) / sample_rate

        print()
        print("=" * 50)
        print("✅ AUDIO GÉNÉRÉ")
        print("=" * 50)
        print(f"   Fichier : {output_path.name}")
        print(f"   Durée   : {duration:.2f} secondes")
        print(f"   Sample rate : {sample_rate} Hz")
        print("=" * 50)

        return str(output_path), duration


# Libérer mémoire et créer le module
clear_gpu_memory()

tts = TTSModule(model_name="facebook/mms-tts-yor")

print("\n✅ Module TTS prêt !")
```

```python
#@title 🔧 Ajuster la vitesse de l'audio { display-mode: "form" }

import subprocess
from pathlib import Path

# Chemins dynamiques (basés sur le nom de la vidéo)
video_stem = Path(VIDEO_NAME).stem
audio_input = TEMP_DIR / f"{video_stem}_audio_yoruba.wav"
audio_adjusted = TEMP_DIR / f"{video_stem}_audio_adjusted.wav"

# Calculer le facteur de vitesse
# audio > vidéo → accélérer (facteur > 1)
# audio < vidéo → ralentir (facteur < 1)
speed_factor = audio_duration / video_duration

print(f"📊 Ajustement de la vitesse :")
print(f"   Audio original : {audio_duration:.2f} sec")
print(f"   Vidéo cible    : {video_duration:.2f} sec")
print(f"   Facteur        : {speed_factor:.3f}")

if speed_factor > 1:
    print(f"   → Accélération de {(speed_factor - 1) * 100:.0f}%")
else:
    print(f"   → Ralentissement de {(1 - speed_factor) * 100:.0f}%")
print()

# FFmpeg atempo accepte entre 0.5 et 2.0
# Si hors limites, on enchaîne plusieurs atempo
if speed_factor > 2.0:
    # Enchaîner deux atempo
    factor1 = 2.0
    factor2 = speed_factor / 2.0
    filter_str = f"atempo={factor1},atempo={factor2}"
    print(f"   ⚠️ Facteur > 2.0, utilisation de deux filtres")
elif speed_factor < 0.5:
    factor1 = 0.5
    factor2 = speed_factor / 0.5
    filter_str = f"atempo={factor1},atempo={factor2}"
    print(f"   ⚠️ Facteur < 0.5, utilisation de deux filtres")
else:
    filter_str = f"atempo={speed_factor}"

cmd = [
    'ffmpeg', '-y',
    '-i', str(audio_input),
    '-filter:a', filter_str,
    str(audio_adjusted)
]

print("⏳ Ajustement en cours...")
ffmpeg_result = subprocess.run(cmd, capture_output=True, text=True)

if ffmpeg_result.returncode == 0:
    # Vérifier la nouvelle durée
    import wave
    with wave.open(str(audio_adjusted), 'r') as wav:
        frames = wav.getnframes()
        rate = wav.getframerate()
        new_duration = frames / rate

    print()
    print("=" * 50)
    print("✅ AUDIO AJUSTÉ")
    print("=" * 50)
    print(f"   Fichier        : {audio_adjusted.name}")
    print(f"   Nouvelle durée : {new_duration:.2f} sec")
    print(f"   Durée cible    : {video_duration:.2f} sec")
    print(f"   Écart restant  : {abs(new_duration - video_duration):.2f} sec")
    print("=" * 50)
else:
    print(f"❌ Erreur FFmpeg : {ffmpeg_result.stderr}")
```

```python
#@title 👄 Module Lip-Sync { display-mode: "form" }

import subprocess
from pathlib import Path
import shutil

class LipSyncModule:
    """
    Module de synchronisation labiale utilisant Wav2Lip.

    Wav2Lip génère des mouvements de lèvres réalistes
    à partir d'une vidéo et d'un audio.
    """

    def __init__(self, wav2lip_path: str = "Wav2Lip"):
        """
        Initialise le module.

        Args:
            wav2lip_path: Chemin vers le dossier Wav2Lip
        """
        self.wav2lip_path = Path(wav2lip_path)
        self.checkpoint_path = self.wav2lip_path / "checkpoints" / "wav2lip_gan.pth"

        # Vérifier que tout est en place
        if not self.checkpoint_path.exists():
            raise FileNotFoundError(f"❌ Modèle non trouvé : {self.checkpoint_path}")

        print("✅ Module Lip-Sync initialisé")

    def synchroniser(self, video_path: str, audio_path: str, output_path: str = None) -> str:
        """
        Synchronise les lèvres d'une vidéo avec un audio.

        Args:
            video_path: Chemin de la vidéo source
            audio_path: Chemin de l'audio à synchroniser
            output_path: Chemin de sortie (optionnel)

        Returns:
            Chemin de la vidéo synchronisée
        """
        video_path = Path(video_path)
        audio_path = Path(audio_path)

        if output_path is None:
            output_path = TEMP_DIR / "video_lipsync.mp4"
        output_path = Path(output_path)

        print("=" * 50)
        print("👄 SYNCHRONISATION LABIALE")
        print("=" * 50)
        print(f"   Vidéo  : {video_path.name}")
        print(f"   Audio  : {audio_path.name}")
        print(f"   Sortie : {output_path.name}")
        print()
        print("⏳ Traitement en cours...")
        print("   (Cela peut prendre 2-5 minutes)")
        print()

        # Commande Wav2Lip
        cmd = [
            "python", "inference.py",
            "--checkpoint_path", str(self.checkpoint_path),
            "--face", str(video_path.absolute()),
            "--audio", str(audio_path.absolute()),
            "--outfile", str(output_path.absolute()),
            "--resize_factor", "1",
            "--nosmooth"
        ]

        # Exécuter depuis le dossier Wav2Lip
        result = subprocess.run(
            cmd,
            cwd=str(self.wav2lip_path),
            capture_output=True,
            text=True
        )

        if result.returncode != 0:
            print("❌ Erreur Wav2Lip :")
            print(result.stderr[-2000:])  # Derniers 2000 caractères
            raise RuntimeError("Lip-sync échoué")

        if output_path.exists():
            print()
            print("=" * 50)
            print("✅ LIP-SYNC TERMINÉ")
            print("=" * 50)
            print(f"   Fichier : {output_path}")

            # Taille du fichier
            size_mb = output_path.stat().st_size / (1024 * 1024)
            print(f"   Taille  : {size_mb:.2f} MB")
            print("=" * 50)

            return str(output_path)
        else:
            raise RuntimeError("❌ Fichier de sortie non créé")


# Créer le module
lipsync = LipSyncModule()
```
