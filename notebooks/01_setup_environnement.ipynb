```python
#@title 🔍 Détection de la plateforme { display-mode: "form" }

import os
import sys

def detect_platform():
    """Détecte automatiquement la plateforme d'exécution."""

    if 'google.colab' in sys.modules:
        return 'COLAB'
    elif 'KAGGLE_KERNEL_RUN_TYPE' in os.environ:
        return 'KAGGLE'
    elif 'LIGHTNING_CLOUDSPACE_HOST' in os.environ:
        return 'LIGHTNING'
    else:
        return 'LOCAL'

PLATFORM = detect_platform()

print(f"🖥️  Plateforme détectée : {PLATFORM}")

# Configurations spécifiques
if PLATFORM == 'COLAB':
    print("   → Google Colab détecté")
    print("   → Drive sera monté pour la persistance")
elif PLATFORM == 'KAGGLE':
    print("   → Kaggle détecté")
    print("   → Les données persistent dans /kaggle/working")
```

```python
#@title 🎮 Vérification du GPU { display-mode: "form" }

import torch

print("=" * 50)
print("🎮 VÉRIFICATION GPU")
print("=" * 50)

if torch.cuda.is_available():
    gpu_name = torch.cuda.get_device_name(0)
    vram_total = torch.cuda.get_device_properties(0).total_memory / 1e9
    vram_free = (torch.cuda.get_device_properties(0).total_memory - torch.cuda.memory_allocated(0)) / 1e9

    print(f"✅ GPU disponible : {gpu_name}")
    print(f"   VRAM totale    : {vram_total:.1f} GB")
    print(f"   VRAM libre     : {vram_free:.1f} GB")

    # Recommandations basées sur la VRAM
    print()
    print("📊 Recommandations basées sur votre GPU :")
    if vram_total >= 15:
        print("   ✅ Whisper large-v3 : OK")
        print("   ✅ NLLB-1.3B : OK")
        print("   ✅ Wav2Lip : OK")
    elif vram_total >= 8:
        print("   ⚠️  Whisper medium recommandé")
        print("   ⚠️  NLLB-600M recommandé")
        print("   ✅ Wav2Lip : OK")
    else:
        print("   ⚠️  Ressources limitées, certains modèles peuvent ne pas fonctionner")
else:
    print("❌ Aucun GPU détecté !")
    print("   → Allez dans : Modifier > Parametres de notebook > GPU")

print("=" * 50)
```

```python
#@title 📁 Montage Google Drive { display-mode: "form" }

if PLATFORM == 'COLAB':
    from google.colab import drive

    print("📁 Montage de Google Drive...")
    drive.mount('/content/drive')

    # Chemins du projet
    DRIVE_ROOT = "/content/drive/MyDrive/dubbing_project"

    # Vérifier que la structure existe
    required_folders = ['inputs', 'outputs', 'temp', 'checkpoints', 'models']

    print()
    print("🔍 Vérification de la structure Drive :")
    all_ok = True
    for folder in required_folders:
        folder_path = f"{DRIVE_ROOT}/{folder}"
        if os.path.exists(folder_path):
            print(f"   ✅ {folder}/")
        else:
            print(f"   ❌ {folder}/ manquant !")
            all_ok = False

    if all_ok:
        print()
        print("✅ Structure Drive OK !")
    else:
        print()
        print("⚠️  Créez les dossiers manquants dans Drive")

elif PLATFORM == 'KAGGLE':
    DRIVE_ROOT = "/kaggle/working/dubbing_project"
    os.makedirs(DRIVE_ROOT, exist_ok=True)
    print(f"📁 Dossier de travail : {DRIVE_ROOT}")
```
